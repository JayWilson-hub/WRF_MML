```python
import numpy as np
import matplotlib.pyplot as plt
import rasterio as rio
%matplotlib inline
```

```python
#carregando as imagens
band10 = rio.open('LC08_L1TP_217076_20170728_20170810_01_T1/LC08_L1TP_217076_20170728_20170810_01_T1_B10.TIF')
band11 = rio.open('LC08_L1TP_217076_20170728_20170810_01_T1/LC08_L1TP_217076_20170728_20170810_01_T1_B11.TIF')
```

```python
plt.matshow(band11.read(1))
```

```python
# convertendo para Reflecancia 
r10 = 0.00033420*band10.read(1)+0.1
r11 = 0.00033420*band11.read(1)+0.1
```

```python
#convertendo para TOA
K1_CONSTANT_BAND_10 = 774.8853
K2_CONSTANT_BAND_10 = 1321.0789
K1_CONSTANT_BAND_11 = 480.8883
K2_CONSTANT_BAND_11 = 1201.1442

```

```python
TOA10 = K2_CONSTANT_BAND_10/(np.log(K1_CONSTANT_BAND_10/r10+1))
```

```python
TOA11 = K2_CONSTANT_BAND_11/(np.log(K1_CONSTANT_BAND_11/r11+1))
```

```python
#convertendo celsius
TOA10 = TOA10 - 273.15
TOA11 = TOA11 - 273.15
```

# Salvando os modelos 

```python
modelo1 = rio.open('LC08_L1TP_217076_20170914_20170928_01_T1/TOA10.tif','w', driver='GTiff',
                  height=TOA10.shape[0],width=TOA10.shape[1],
                  count=1, dtype=TOA10.dtype,
                  crs='EPSG:32723',transform=band10.transform)

modelo2 = rio.open('LC08_L1TP_217076_20170914_20170928_01_T1/TOA11.tif','w', driver='GTiff',
                  height=TOA11.shape[0],width=TOA11.shape[1],
                  count=1, dtype=TOA11.dtype,
                  crs='EPSG:32723',transform=band11.transform)
```

```python
modelo1.write(TOA10,1)
modelo1.close()

modelo2.write(TOA11,1)
modelo2.close()
```
