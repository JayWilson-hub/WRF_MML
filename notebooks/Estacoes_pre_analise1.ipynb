```python
tmax= df.iloc[:,:24].values
tmin=df.iloc[:,24:].values
tmax = tmax.ravel()
tmin = tmin.ravel()
```

```python
file=['../dados_meteorologicos/INMET/MARAMBAIA_01_2015_12_2017.csv',
 '../dados_meteorologicos/INMET/XEREM_01_2015_12_2017.csv', 
 '../dados_meteorologicos/INMET/FORTE_DE_COPACABANA_01_01_2015_31_12_2017.csv']
```

```python
for i in range(len(file)):
    df = pd.read_csv(file[i])
    tmax= df.iloc[:,:24].values
    tmin=df.iloc[:,24:].values
    tmax = tmax.ravel()
    tmin = tmin.ravel()
    dados = pd.DataFrame({'tmin':tmin,'tmax':tmax})
    dados.index = pd.DatetimeIndex(start = '31-12-2014 21:00:00', 
                                     freq = '1H', periods=len(dados))
    globals()[estacoes[i]]= dados
```
